```python
import math
import meep as mp
import numpy as np
import matplotlib.pyplot as plt

resolution = 20   # pixels/um
eps = 2          # dielectric constant of waveguide
w = 20        # width of waveguide
r1 = 181/385/2
r2 = 81/385/2
d = 1.4           # defect spacing (ordinary spacing = 1)
sy = 30     # size of cell in y direction (perpendicular to wvg.)
pad = 2           # padding between last hole and PML edge
dpml = 1          # PML thickness
sx =30 # size of cell in x direction
cell = mp.Vector3(sx,sy,0)#计算单元
nSi = 3.47
Si = mp.Medium(index=nSi)
blk = mp.Block(size=mp.Vector3(17,14,mp.inf), center=mp.Vector3(0,0,0),
               material=Si)
geometry = [blk]
geometry.append(mp.Block(size=mp.Vector3(12,4,mp.inf), center=mp.Vector3(-12.5,0,0),
               material=Si))
geometry.append(mp.Block(size=mp.Vector3(12,4,mp.inf), center=mp.Vector3(12.5,0,0),
               material=Si))
a = 1/math.sqrt(3)
b = 1/2
c = 3*a/2
#colomn 1 btm
for i in range(9):
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(0+i, 0), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(0+i, a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(0+i,3*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(0+i, 4*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(0+i, 6*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(0+i, 7*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(0+i, 9*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(0+i, 10*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(0+i, 12*a), material=mp.air))
#colomn 1 up
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(0+i, -2*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(0+i, -3*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(0+i, -5*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(0+i, -6*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(0+i, -8*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(0+i, -9*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(0+i, -11*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(0+i, -12*a), material=mp.air))
#colomn 2 btm
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(b+i, c), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(b+i, c+a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(b+i, c+3*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(b+i, c+4*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(b+i, c+6*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(b+i, c+7*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(b+i, c+9*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(b+i, c+10*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(b+i, c+12*a), material=mp.air))
#colomn 2 up
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(b+i, c-2*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(b+i, c-3*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(b+i, c-5*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(b+i, c-6*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(b+i, c-8*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(b+i, c-9*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(b+i, c-11*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(b+i, c-12*a), material=mp.air))
    
    
    
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(0+i), 0), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(0+i), a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(0+i),3*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(0+i), 4*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(0+i), 6*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(0+i), 7*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(0+i), 9*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(0+i), 10*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(0+i), 12*a), material=mp.air))
#colomn 1 up
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(0+i), -2*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(0+i), -3*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(0+i), -5*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(0+i), -6*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(0+i), -8*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(0+i), -9*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(0+i), -11*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(0+i), -12*a), material=mp.air))
#colomn 2 btm
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(b+i), c), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(b+i), c+a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(b+i), c+3*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(b+i), c+4*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(b+i), c+6*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(b+i), c+7*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(b+i), c+9*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(b+i), c+10*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(b+i), c+12*a), material=mp.air))
#colomn 2 up
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(b+i), c-2*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(b+i), c-3*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(b+i), c-5*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(b+i), c-6*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(b+i), c-8*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(b+i), c-9*a), material=mp.air))
    geometry.append(mp.Cylinder(r1, center=mp.Vector3(-(b+i), c-11*a), material=mp.air))
    geometry.append(mp.Cylinder(r2, center=mp.Vector3(-(b+i), c-12*a), material=mp.air))

pml_layers = [mp.PML(1.0)]
fsrc = 0.7
kx = 1    # initial guess for wavevector in x-direction of eigenmode
k_points = mp.Vector3(0, 1/2)

fsrc =2/1.430
sources = [mp.EigenModeSource(src=mp.ContinuousSource(fsrc),
                               direction=mp.X,
                               component=mp.Ez,
                                amplitude=1,
                                center=mp.Vector3(-10,0,0),
                                size=mp.Vector3(0,5,0),
                              eig_band=1,
                               eig_kpoint=k_points,
                                eig_parity=mp.ODD_Y,
                               eig_match_freq=True)]
sim = mp.Simulation(cell_size=cell,
                    resolution=resolution,
                    boundary_layers=pml_layers,
                    sources=sources,
                    geometry=geometry
                   )
sim.run(until=200)
```

```python
import numpy as np
import matplotlib.pyplot as plt
eps_data = sim.get_array(center=mp.Vector3(), size=cell, component=mp.Dielectric)
plt.figure()
plt.imshow(eps_data.transpose(), interpolation='spline36', cmap='binary')
plt.axis('off')
plt.show()
ez_data = sim.get_array(center=mp.Vector3(), size=cell, component=mp.Ez)
plt.figure()
plt.imshow(eps_data.transpose(), interpolation='spline36', cmap='binary')
plt.imshow(ez_data.transpose(), interpolation='spline36', cmap='RdBu', alpha=0.9)
plt.axis('off')
plt.show()
```
